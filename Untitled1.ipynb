```python
x_train,x_test,y_train,y_test = train_test_split(data, labels, test_size=0.2, shuffle=True, random_state=7)
```

```python
from tensorflow.keras.applications.xception import Xception
from tensorflow.keras.utils import plot_model

from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dropout
```

```python
model = tf.keras.applications.efficientnet.EfficientNetB4(
    include_top=False,
    weights="imagenet",
    input_tensor=None,
    input_shape=(128,128,3),
    pooling='avg',
    classes=2,
)
for layers in model.layers:
    layers.trainable=False
    
flat1 = keras.layers.Flatten()(model.layers[-1].output)
class1 = keras.layers.Dense(256, activation='relu')(flat1)
class1 = keras.layers.Dropout(0.5)(class1)
output = keras.layers.Dense(1, activation='sigmoid')(class1)
model=keras.Model(inputs=model.input, outputs=output)
```

```python
from keras import backend as K

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

callback = tf.keras.callbacks.EarlyStopping(
    monitor="val_f1_m",
    patience=5,
    verbose=0,
    mode="max",
    baseline=None,
)

# fit the model
history = model.fit(
    x = x_train,
    y = y_train,
    validation_split= 0.2,
    batch_size = 64,
    epochs=50,
    callbacks=[callback],
    verbose=(2),
)

# evaluate the model
scores = model.evaluate(x_test, y_test, verbose=1)
print("Test Loss: ", scores[0])
print("Test Accuracy: ", scores[1])
```
